# Reflex Notebooks Quickstart

Turn this notebook into a deployed Reflex app in three lines of Python.

```python
import reflex as rx

rx.notebook()  # one-time init: registers IPython hooks
```

```python
import pandas as pd

df = pd.DataFrame({
    "category": ["A", "B", "A", "C", "B", "A"],
    "value": [10, 25, 30, 5, 17, 22],
})
df
```

```python
category = rx.notebook.select(
    sorted(df["category"].unique()),
    label="Category",
)
threshold = rx.notebook.slider(0, 50, default=10, label="Min value")
```

```python
filtered = df[(df["category"] == category) & (df["value"] >= threshold)]
rx.notebook.display(filtered)
```

```python
# Inspect the Reflex source the notebook would compile to:
rx.notebook.view_source(app_name="sales_demo")
```

```python
# Materialize the app on disk and print the run command:
rx.notebook.deploy(app_name="sales_demo")
```
