# Sales Dashboard вЂ” Reflex Notebooks Sample

A single-cell dashboard built on the Reflex Notebooks MVP. The filter widgets
live at the top of the dashboard cell, and the filtered table, headline metric,
and charts render directly underneath them. Move any widget and every output
below it updates in place inside the same cell.

## Setup

```python
import datetime as dt
import random

import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
from IPython.display import Markdown

import reflex as rx

rx.notebook()

random.seed(7)
regions = ["North", "South", "East", "West"]
products = ["Widget", "Gadget", "Sprocket", "Gizmo"]
customers = ["Acme", "Globex", "Initech", "Umbrella", "Soylent", "Stark"]

start = dt.date(2025, 1, 1)
sales = pd.DataFrame([
    {
        "date": start + dt.timedelta(days=i),
        "region": random.choice(regions),
        "product": random.choice(products),
        "customer": random.choice(customers),
        "units": random.randint(1, 20),
        "revenue": round(random.uniform(50, 1500), 2),
    }
    for i in range(120)
])
```

## Dashboard

```python
region = rx.notebook.select(
    ["All", *sorted(sales["region"].unique())], default="All", label="Region"
)
product = rx.notebook.select(
    ["All", *sorted(sales["product"].unique())], default="All", label="Product"
)
min_revenue = rx.notebook.slider(0, 1500, default=200, step=50, label="Min revenue ($)")
after_date = rx.notebook.date_picker(
    default=dt.date(2025, 2, 1), label="Sales on/after"
)
customer_query = rx.notebook.text_input(
    default="", label="Customer contains", placeholder="e.g. Acme"
)
show_totals = rx.notebook.checkbox(default=True, label="Show totals")

mask = (sales["revenue"] >= min_revenue) & (sales["date"] >= after_date)
if region != "All":
    mask &= sales["region"] == region
if product != "All":
    mask &= sales["product"] == product
if customer_query:
    mask &= sales["customer"].str.contains(customer_query, case=False, na=False)
filtered = sales[mask].sort_values("date").reset_index(drop=True)

if show_totals:
    total = float(filtered["revenue"].sum())
    rx.notebook.display(
        Markdown(
            f"### Total revenue: **${total:,.2f}**  \n"
            f"Across **{len(filtered)}** rows matching the current filter."
        )
    )

rx.notebook.display(filtered.head(15))

by_region = filtered.groupby("region")["revenue"].sum().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(6, 3))
by_region.plot(kind="bar", ax=ax, color="#5B7CFA")
ax.set_title("Revenue by region")
ax.set_ylabel("Revenue ($)")
ax.set_xlabel("")
fig.tight_layout()
rx.notebook.display(fig)

daily = filtered.groupby("date")[["revenue"]].sum().reset_index()
fig_px = px.line(daily, x="date", y="revenue", markers=True, title="Daily revenue")
fig_px.update_layout(height=320, margin={"l": 20, "r": 20, "t": 40, "b": 20})
rx.notebook.display(fig_px)

top_customer = (
    filtered.groupby("customer")["revenue"].sum().sort_values(ascending=False)
)
if len(top_customer):
    leader, leader_revenue = top_customer.index[0], top_customer.iloc[0]
    rx.notebook.display(
        Markdown(f"**Top customer:** `{leader}` at **${leader_revenue:,.2f}**.")
    )
else:
    rx.notebook.display(Markdown("_No rows matched the current filter._"))
```

## Random row picker

```python
clicked = rx.notebook.button(label="Pick a random row")
if clicked and len(filtered):
    rx.notebook.display(filtered.sample(1).reset_index(drop=True))
```
